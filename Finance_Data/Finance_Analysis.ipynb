## 2. Preferência de investimento

Essas análises nos ajudarão a entender as preferências de investimento dos clientes e as relações entre as diferentes opções de investimento. A análise exploratória nos ajuda a identificar padrões e relações iniciais nos dados, permitindo-nos formular hipóteses para investigações mais aprofundadas.

```python
# Criando modelo de contagem de investimentos em cada categoria

# Criando uma lista contendo os nomes das diferentes categorias de investimento no DataFrame
investimentos = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']

# Selecionando apenas as colunas de investimento do DataFrame 'df_fd' usando 'df_fd[investimentos]', o que nos dá um novo DataFrame contendo apenas essas colunas
# Em seguida incluo o método 'sum()' para calcular a soma ao longo das colunas, o que nos dá a contagem total de investidores em cada categoria
contagem_investimentos = df_fd[investimentos].sum()
print(contagem_investimentos)
```

```python
# Plotando um gráfico de barras para visualizar a contagem de investimentos em cada categoria
plt.figure(figsize=(10, 6))
contagem_investimentos.sort_values().plot(kind='barh', color='salmon')
plt.title('Preferências de Investimento dos Clientes')
plt.xlabel('Número de Investidores')
plt.ylabel('Categoria de Investimento')
plt.show()
```

Com base na contagem de investimentos em cada categoria, obtive alguns insights valiosos: 

1. Popularidade de investimentos de baixo risco: foi identificado que investimentos de baixo risco, como debentures, government bonds e fixed deposits, têm contagens relativamente altas, isso sugere que uma parte significativa dos investidores pode preferir opções de investimento mais seguras e estáveis;

2. Diversificação de portfólio: a diversificação de portfólio é uma estratégia comum para reduzir o risco. A presença de investimentos em diferentes categorias, como Equity Market, PPF e Gold, indica que os investidores podem estar buscando diversificar seus portfólios para equilibrar o risco e o retorno;

3. Interesse em Crescimento e Renda Passiva: investimentos como Equity Market e Gold, podem indicar um interesse dos investidores em buscar crescimento e retornos mais altos ao longo do tempo. Por outro lado, opções como Fixed Deposits e Government Bonds, podem representar uma preferência por renda passiva e retornos estáveis;

4. Preferências pessoais e objetivos de investimento: a distribuição dos investimentos pode refletir as preferências pessoais e os objetivos de investimento únicos de cada investidor. Alguns podem priorizar a segurança e a estabilidade enquanto outros podem buscar retornos mais altos e estão dispostos a assumir mais riscos;

5. Sazonalidade e tendências de mercado: mudanças na contagem de investimentos em diferentes categorias ao longo do tempo, podem refletir sazonalidade ou tendências do mercado. Por exemplo, um aumento na contagem de investimentos em Equity Market pode ocorrer durante períodos de otimismo do mercado.

```python
# Criando modelo de correlação entre as diferentes opções de investimento
correlacao_investimentos = df_fd[investimentos].corr()

# Plotando um mapa de calor da matriz de correlação
plt.figure(figsize=(10,8))
sns.heatmap(correlacao_investimentos, annot=True, cmap='pink', fmt=".2f")
plt.title('Correlação entre Opções de Investimento')
plt.show()
```

Com base na matriz de correlação entre as diferentes opções de investimento, podemos obter os seguintes insights:

1. Correlação Positiva Moderada entre Mutual_Funds e Equity_Market:

A correlação de aproximadamente 0.33 entre Mutual_Funds e Equity_Market indica uma relação moderadamente positiva. Isso sugere que os movimentos nos fundos mútuos estão moderadamente alinhados com os movimentos no mercado de ações.

2. Correlação Negativa Moderada entre Mutual_Funds e Debentures:

A correlação de aproximadamente -0.35 entre Mutual_Funds e Debentures sugere uma relação negativa moderada. Isso implica que os investidores podem ter uma tendência a investir menos em debêntures quando investem mais em fundos mútuos, e vice-versa.

3. Correlação Negativa Moderada entre Mutual_Funds e Government_Bonds:

A correlação de aproximadamente -0.11 entre Mutual_Funds e Government_Bonds indica uma relação negativa moderada. Isso sugere que os investidores podem preferir investir menos em fundos mútuos quando investem mais em títulos do governo, e vice-versa.

4. Correlação Negativa Forte entre Fixed_Deposits e Outros Investimentos:

As correlações negativas mais fortes de aproximadamente -0.53 entre Fixed_Deposits e Government_Bonds, e de aproximadamente -0.47 entre Fixed_Deposits e Debentures, indicam uma preferência por investimentos seguros como depósitos fixos em detrimento de investimentos de maior risco, como títulos do governo e debêntures.

5. Correlação Negativa Moderada entre PPF e Equity_Market:

A correlação de aproximadamente -0.43 entre PPF e Equity_Market sugere uma relação negativa moderada. Isso indica que os investidores podem preferir investir menos em PPF quando investem mais no mercado de ações, e vice-versa.

6. Correlação Negativa Moderada entre Gold e Mutual_Funds, Equity_Market, e Debentures:

As correlações negativas moderadas entre Gold e os outros investimentos sugerem que os investidores podem ver o ouro como uma alternativa de investimento em momentos de menor confiança no mercado de ações ou em investimentos de renda fixa.


Com base nos insights derivados da análise da matriz de correlação entre os diferentes tipos de investimento, podemos concluir que a diversificação é crucial para os investidores, pois distribuir seus investimentos entre diferentes categorias de ativos pode ajudar a reduzir o risco total da carteira. Além disso, a correlação negativa entre investimentos mais seguros, como títulos do governo e depósitos fixos, e investimentos de maior risco, como ações e fundos mútuos, sugere que os investidores buscam um equilíbrio entre risco e segurança. A correlação negativa moderada entre ouro e outros investimentos indica que o ouro pode ser considerado uma alternativa de investimento em períodos de volatilidade nos mercados financeiros. Portanto, uma alocação de ativos consciente e o monitoramento regular da carteira são essenciais para os investidores alcançarem seus objetivos financeiros de longo prazo.

## 4. Perfil e Hábitos Financeiros dos Jovens Adultos (24-31 anos)

Agora que identificamos um público alvo, podemos entender quais são seus hábitos financeiros dos jovens adultos na faixa etária de 24 a 31 anos, entendendo como adaptar melhor os produtos e serviços.

```python
df_fd.head()
```

```python
# Filtrando os dados para incluir apenas os correspondentes com idades entre 24 a 31 anos
df_filtered = df_fd[(df_fd['age']>= 24) & (df_fd['age'] <= 31)]
print(df_filtered)
```

```python
df_filtered.head()
```

```python
# Realiando uma análise descritiva das variáveis relacionadas aos hábitos financeiros
finance_habits_desc = df_filtered.describe(include='all')
print(finance_habits_desc)
```

Agora vamos criar um modelo de visualização para explorar os hábitos financeiros dos respondentes nesta faixa etária.

```python

# Criando modelo de visualização em gráfico de barras para preferências de investimentos
plt.figure(figsize=(10,6))
sns.countplot(x='Investment_Avenues',data=df_filtered)
plt.title('Preferências de Investimento dos Respondentes (24-31 anos)')
plt.xlabel('Opções de Investimento')
plt.ylabel('Número de Respondentes')
plt.show()
```

```python
df_filtered.head()
```

```python
# Calcular a contagem de investimentos em cada categoria
contagem_investimentos = df_filtered[['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']].sum()

# Plotar um gráfico de barras para visualizar as preferências de investimento
plt.figure(figsize=(10, 6))
contagem_investimentos.sort_values().plot(kind='barh', color='salmon')
plt.title('Preferências de Investimento dos Jovens (24-31 anos)')
plt.xlabel('Número de Investidores')
plt.ylabel('Categoria de Investimento')
plt.show()
```

Algumas observações:

- Em geral, o número de investidores em cada categoria é maior no grupo geral em comparação com o grupo de investidores entre 24 e 31 anos.
- Ouro (Gold) é a categoria com o maior número de investidores em ambos os grupos.
- Os fundos mútuos (Mutual Funds) e o mercado de ações (Equity Market) são as próximas opções mais populares em ambos os grupos, seguidos por debêntures e títulos do governo.
- As categorias com menor número de investidores em ambos os grupos são PPF (Plano de Providência para Aposentadoria) e Depósitos Fixos (Fixed Deposits).
